##### Sélection des indicateurs

```python
indic = Data_df.groupby(['Indicator Name']).count().sort_values(by=['2010'],ascending = False).reset_index()[['Indicator Name','2010']]
indic
```

On va choisir les quatre premiers indicateurs qui représentent :<br>
– « Population, total » : (population totale)<br>
– « Population growth (annual %) » : (Croissance de la population)<br>
– « GDP at market prices (current US\\$) » : (PIB au prix du marché)<br>
– « GDP per capita (current US\\$) » : (PIB par habitant)<br>

Et « Internet users (per 100 people) » : (Utilisateurs d’internet)

```python
select_indic = indic["Indicator Name"][:4].tolist()
select_indic.extend(["Internet users (per 100 people)", "GNI (current US$)"])
```

Maintenant, on va sélectionner les indicateurs qui comportent les mots : « secondary », « tertiary » et « population ».

```python
indic_search = [c for c in indic['Indicator Name'].str.lower()
               if(("secondary" in c) | ("tertiary" in c) | ("population," in c) ) ]
search = (indic[indic["Indicator Name"]\
                   .str.lower().isin(indic_search)]\
                   .loc[indic["2010"]>100,:]
             )
search
```

On retire les indicateurs qui comportent : « female », « male » et « wittgenstein ».

```python
indic_supp = [c for c in indic['Indicator Name'].str.lower()
               if(("female" in c) | ("male" in c) ) ]
supp = (indic[indic["Indicator Name"]\
                   .str.lower().isin(indic_supp)]\
                   .loc[indic["2010"]>150,:]
             )
```

```python
search[~search['Indicator Name'].isin(supp['Indicator Name'])]
```

On va choisir ces indicateurs :

```python
select_indic.extend(
["Population of the official age for secondary education, both sexes (number)",
 "Population of the official age for tertiary education, both sexes (number)",
 "Enrolment in secondary general, both sexes (number)",
 "Enrolment in tertiary education, all programmes, both sexes (number)",
 "Population, ages 15-64, total",
 "Population, ages 13-19, total"]
)
select_indic
```

## Calcul du Score

```python
# Nombre de pays
len(data_ref)
```

```python
ind_score = ["Enrolment in secondary general, both sexes (number)", "Enrolment in tertiary education, all programmes, both sexes (number)",
               "GDP at market prices (current US$)", "GDP per capita (current US$)",
               "GNI (current US$)", "Population growth (annual %)"]
```

```python
#Calcul du score
data_score = data_ref
new_cols = [c+"_norm" for c in ind_score]
for indic in ind_score :
    data_score[indic+"_norm"] = (data_score[indic] - min(data_score[indic])) / (max(data_score[indic])- min(data_score[indic]))*100.
    
data_score["score"] = round(data_score[new_cols].mean(axis=1, skipna=True ), 0) 
data_score = data_score.sort_values( by=["score"], ascending=False ).reset_index(drop=True)
data_score = data_score[["Country Name", "score"]]
```

```python
data_score.head(20)
```

```python
data_score.to_csv('data_score.csv')
```

```python
data_pays = data_ref[data_ref['Country Name'].isin(data_score['Country Name'])]
data_pays = data_pays.sort_values(by ='score', ascending = False)
data_pays
```

```python
data_pays.to_csv('data_pays.csv')
```
